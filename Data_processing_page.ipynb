```python
from faker import Faker
import random
import pandas as pd
import numpy as np
import datetime
```

```python
fake = Faker()
```

```python
data = []

# Common hashtags
common_hashtags = ['#instagood', '#photooftheday', '#beautiful', '#happy', '#cute', '#tbt', '#picoftheday', '#selfie']

# Common comments
common_comments = ['Looks homeless', 'you are a whiner', 'So cute!', 'Looks like your voice is putting me to sleep', 'offensive', 'Fantastic!', 'Nice shot!', 'Beautiful!', 'Good try, don\'t try again.', 'GOAT', 'More posts like this', 'Can do better', 'didn\'t pass the vibe check', 'Such a L', 'Gives me the Ick', 'Check your DM!!']

for _ in range(100):
    instagram_handle = fake.user_name() # fake username
    regular_name = fake.name() # fake name
    num_pictures = random.randint(0, 20) # number of pictures posted
    stories_posted = random.randint(0, 50) # number of stories posted
    descriptions = [fake.sentence() for _ in range(num_pictures)] # descriptions of the pictures
    captions = [fake.sentence() for _ in range(num_pictures)] # captions under images
    hashtags = [random.sample(common_hashtags, random.randint(1, 5)) for _ in range(num_pictures)]  # Generate hashtags for each post
    num_tags_by_you = [random.randint(0, 5) for _ in range(num_pictures)] # pictures tagged by you
    num_tags_to_you = [random.randint(0, 5) for _ in range(num_pictures)] # pictures tagged to you
    comments_on_other_posts = [random.sample(common_comments, random.randint(0, 5)) for _ in range(num_pictures)] # Generate comments left by the person on other people's posts
    comments_on_own_posts = [random.sample(common_comments, random.randint(0, 10)) for _ in range(num_pictures)] # Generate comments on the person's own posts
    followers = random.randint(0, 10000) # followers
    following = random.randint(0, 1000) # following
    interactions_with_friends = random.randint(0, 50) # shared posts with your followers
     # Determine number of close friends
    if following == 0:
        close_friends = 0
    else:
        close_friends = random.randint(0, min(followers, following))
    interactions_with_close_friends = random.randint(0, close_friends) # Determine interactions with close friends
    data.append([instagram_handle,regular_name, num_pictures, stories_posted, descriptions, captions, hashtags, num_tags_by_you, num_tags_to_you, comments_on_other_posts, comments_on_own_posts, followers, following, interactions_with_friends, close_friends, interactions_with_close_friends]) # putting it all together
```

```python
# Create DataFrame
columns = ['Instagram Handle', 'Name', 'Number of Pictures', 'Stories Posted', 'Descriptions', 'Captions', 'Hastags', 'Number of Tags by You', 'Number of Tags to You', 'Comments on Others', 'Comments on post',  'Followers', 'Following','Interactions with Friends', 'Close Friends', 'Interactions with Close Friends']
df = pd.DataFrame(data, columns=columns)
```

```python
# Generate weekly data for 4 weeks
weekly_data = []
start_date = datetime.date.today() - datetime.timedelta(weeks=4)
for i in range(4):
    end_date = start_date + datetime.timedelta(days=6)
    weekly_data.append([start_date.strftime('%Y-%m-%d'), end_date.strftime('%Y-%m-%d')])
    start_date = end_date + datetime.timedelta(days=1)
```

```python
weekly_df = pd.DataFrame(weekly_data, columns=['Start Date', 'End Date']) # Create a DataFrame for weekly data
```

```python
# Concatenate weekly data with person data
weekly_data_all = pd.concat([pd.concat([weekly_df]*len(df), ignore_index=True), pd.concat([df]*4, ignore_index=True)], axis=1)
```

```python
# Rename columns for clarity
weekly_data_all.columns = [f'{col} ({i % 4 + 1})' if i >= 2 else col for i, col in enumerate(weekly_data_all.columns)]
```

```python
weekly_data_all.to_csv('datasets/weekly_instagram.csv', mode='w')
```
