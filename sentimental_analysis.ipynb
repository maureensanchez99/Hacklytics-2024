```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import numpy as np
```

```python
# instance for sentimental analyzation
sentimental_analyzer = SentimentIntensityAnalyzer()
```

```python
instagram_information = pd.read_csv('datasets/tabllm-batch-generate-65c7a1cb0c70a4ebdc0f4bef-data.csv')
instagram_information
```

```python
# get the csv file
comments_and_concerns = pd.read_csv('datasets/comment_db.csv')
# comments_and_concerns = comments_and_concerns.drop(columns=['user_id', 'sentiment'])
comments_and_concerns
```

```python
# drop the id column
instagram_information = instagram_information.drop(columns='user number')
instagram_information
```

```python
instagram_information['Number of Tags by You (2)'] = instagram_information['Number of Tags by You (2)'].apply(len)
instagram_information['Number of Tags to You (3)'] = instagram_information['Number of Tags to You (3)'].apply(len)# 
```

```python
num_comments_per_row_col1 = np.random.randint(1, 10, size=len(instagram_information))  # Generate between 1 and 3 comments for column 1
num_comments_per_row_col2 = np.random.randint(1, 10, size=len(instagram_information))  # Generate between 1 and 3 comments for column 2
```

```python
comments_per_row_col1 = []
comments_per_row_col2 = []
polarity_scores_per_row_col1 = []
polarity_scores_per_row_col2 = []
```

```python
for num_comments_col1, num_comments_col2 in zip(num_comments_per_row_col1, num_comments_per_row_col2):
    random_indices_col1 = np.random.choice(len(comments_and_concerns), size=num_comments_col1, replace=False)  # Select random comments for column 1
    random_indices_col2 = np.random.choice(len(comments_and_concerns), size=num_comments_col2, replace=False)  # Select random comments for column 2
    comments_col1 = [comments_and_concerns.iloc[random_idx]['comments'] for random_idx in random_indices_col1]
    comments_col2 = [comments_and_concerns.iloc[random_idx]['comments'] for random_idx in random_indices_col2]
    # Calculate polarity scores for each comment using VADER
    polarity_scores_col1 = [sentimental_analyzer.polarity_scores(comment)['compound'] for comment in comments_col1]
    polarity_scores_col2 = [sentimental_analyzer.polarity_scores(comment)['compound'] for comment in comments_col2]

    comments_per_row_col1.append(comments_col1)
    comments_per_row_col2.append(comments_col2)
    polarity_scores_per_row_col1.append(polarity_scores_col1)
    polarity_scores_per_row_col2.append(polarity_scores_col2)
```

```python
comments_df_col1 = pd.DataFrame(comments_per_row_col1, columns=[f'Comments_by_you_{i}' for i in range(1, max(num_comments_per_row_col1) + 1)])
comments_df_col1.fillna('None', inplace=True)
comments_df_col1['All Comments by You'] = comments_df_col1.apply(lambda row: [comment if comment != 'None' else 'None' for comment in row], axis=1)

comments_df_col2 = pd.DataFrame(comments_per_row_col2, columns=[f'Comments_to_you_{i}' for i in range(1, max(num_comments_per_row_col2) + 1)])
comments_df_col2.fillna('None', inplace=True)
comments_df_col2['All Comments to You'] = comments_df_col2.apply(lambda row: [comment if comment != 'None' else 'None' for comment in row], axis=1)
```

```python
# Create DataFrames from the generated polarity scores lists
polarity_scores_df_col1 = pd.DataFrame(polarity_scores_per_row_col1, columns=[f'Polarity_by_you_{i}' for i in range(1, max(num_comments_per_row_col1) + 1)])
polarity_scores_df_col1.fillna(0, inplace=True)
polarity_scores_df_col1['BY_score'] = polarity_scores_df_col1.apply(lambda row: [comment if comment != 0 else 0 for comment in row], axis=1)

polarity_scores_df_col2 = pd.DataFrame(polarity_scores_per_row_col2, columns=[f'Polarity_to_you_{i}' for i in range(1, max(num_comments_per_row_col2) + 1)])
polarity_scores_df_col2.fillna(0, inplace=True)
polarity_scores_df_col2['TO_score'] = polarity_scores_df_col2.apply(lambda row: [comment if comment != 0 else 0 for comment in row], axis=1)
```

```python
instagram_information = pd.concat([instagram_information, comments_df_col1['All Comments by You'], comments_df_col2['All Comments to You'], polarity_scores_df_col1['BY_score'], polarity_scores_df_col2['TO_score']], axis=1)
```

```python
# Function to calculate sentiment score using VADER
def calculate_sentiment_score(description):
    sentences = description.split(', ')
    sentiment_scores = [sentimental_analyzer.polarity_scores(sentence)['compound'] for sentence in sentences]
    return sentiment_scores
```

```python
# Apply sentiment analysis function to each description and store scores in a new column
instagram_information['description_score'] = instagram_information['Descriptions (3)'].apply(calculate_sentiment_score)
```

```python
# Apply sentiment analysis function to each caption and store scores in a new column
instagram_information['caption_Score'] = instagram_information['Captions (4)'].apply(calculate_sentiment_score)
```

```python

from collections import Counter
import re



# Remove brackets and parentheses
instaClean = instagram_information.applymap(lambda x: re.sub(r'[\[\]\(\)"\']', '', str(x)))

# Convert to lowercase
instaClean = instaClean.applymap(lambda x: x.lower())

# Define filler words to remove
filler_words = ['the', 'and', 'is', 'of', 'in', 'to', 'a', 'for', 'on', 'with', 'I','you','your','youre','he','she','we','they','it','its']

# Remove filler words
instaClean = instaClean.applymap(lambda x: ' '.join([word for word in x.split() if word not in filler_words]))



# Group the dataframe by the first column
grouped_df = instaClean.groupby(instaClean.columns[0])


# Iterate over each group
for user, group in grouped_df:
    print(f"User: {user}")
    
    # Concatenate the words for each user
    group['concatenated_words'] = group.apply(lambda row: ' '.join(row[6:8].astype(str)).join(row[11:12].astype(str)).join(row[18:19].astype(str)), axis=1)
    
    # Split the concatenated words into individual words
    words = group['concatenated_words'].str.split()
    
    # Flatten the list of words
    words_flat = [word for sublist in words for word in sublist]
    
    # Count the occurrences of each word
    word_counts = Counter(words_flat)
    
    # Get the 5 most common words
    most_common_words = word_counts.most_common(5)
    
    # Print the most common words for each user
    for word, count in most_common_words:
        print(word, count)
    print()
```
